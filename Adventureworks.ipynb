```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

Sales = pd.read_csv('/content/AdventureWorks Sales Data 2021.csv')
Sales.head()
```

```python
Sales.info()
```

```python
# Set the ordate date data type into datetime
Sales['OrderDate'] = pd.to_datetime(Sales['OrderDate'])
```

```python
product = pd.read_csv('/AdventureWorks Product Lookup.csv')
product.head()
```

```python
# create a product name column in sales table by mapping the product id to the product id of the product table
Sales['ProductName'] = Sales['ProductKey'].map(product.set_index('ProductKey')['ProductName'])
Sales.head()
```

```python
Sales['UnitPrice'] = Sales['ProductKey'].map(product.set_index('ProductKey')['ProductPrice'])
```

```python
#Now add the information related to territory based on the territory lookup table
Territory = pd.read_csv('/AdventureWorks Territory Lookup.csv')
Sales['Region'] = Sales['TerritoryKey'].map(Territory.set_index('SalesTerritoryKey')['Region'])
Sales['Country'] = Sales['TerritoryKey'].map(Territory.set_index('SalesTerritoryKey')['Country'])
```

```python
#The main dataframe should include all columns from the sales table except stockdate, product key, territory key, orderline
df_sales = Sales.drop(['StockDate', 'ProductKey', 'TerritoryKey', 'OrderLineItem'], axis=1)
df_sales.head()
```

```python
df_sales['Revenue']= df_sales['UnitPrice'] * df_sales['OrderQuantity']
df_sales.head()
```

```python
df_sales.to_csv('sales2021.csv')
```
